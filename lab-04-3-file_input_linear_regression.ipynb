## 1. import 

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
```

## 2. random seed

```python
torch.manual_seed(777)
```

## 3. 파일 읽어 들이기, Variable 선언

```python
xy = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float64)
x_data = xy[:, 0:-1]
y_data = xy[:, -1]

print(x_data.shape, '\n', x_data, '\n', len(x_data), '\n')
print(y_data.shape, '\n', y_data)

X = Variable(torch.Tensor(x_data))
Y = Variable(torch.Tensor(y_data))
```

## 4. hypothesis, cost function, optimizer 만들기 

```python
# 입력하는 feature의 개수는 3개이고, 출력되어야 하는 feature의 개수는 1개이기 때문에 아래와 같이 hypothesis를 만든다.
hypothesis = nn.Linear(3, 1, bias=True)
cost_func = nn.MSELoss()
optimizer = torch.optim.SGD(hypothesis.parameters(), lr=1e-5)
```

## 5. Model 훈련시키기

```python
for step in range(20001):
    optimizer.zero_grad()
    
    Y_hat = hypothesis(X)
    cost = cost_func(Y_hat, Y)
    cost.backward()
    optimizer.step()
    
    if step % 1000 == 0:
        print(step, "Cost: ", cost.data.numpy(), "\nPrediction:\n", Y_hat.data.numpy())
```

## 6. Model Test

```python
print("Your score will be", hypothesis(Variable(torch.Tensor([[100, 70, 101]]))).data.numpy())
print("Other scores will be", hypothesis(Variable(torch.Tensor([[60, 70, 110], [90, 100, 80]]))).data.numpy())
```
